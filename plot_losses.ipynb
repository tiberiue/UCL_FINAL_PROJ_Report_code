```python
import pandas as pd
import matplotlib.pyplot as plt
def plot_metrics(infile,model_name):
    metrics = pd.read_csv(infile)
    plt.figure(figsize=(10, 8))
    plt.plot(metrics.index,metrics.Train_Loss_clsf, label='Train')
    plt.plot(metrics.index,metrics.Val_loss_Class,label='Validation')
    plt.ylabel('Loss_clsf')
    plt.xlabel('Epochs')
    plt.legend()
    plt.show() 
    
    plt.figure(figsize=(10, 8))
    plt.plot(metrics.index,metrics.Train_Loss_adv, label='Train')
    plt.plot(metrics.index,metrics.Val_Loss_Adv,label='Validation')
    plt.ylabel('Loss_Adv')
    plt.xlabel('Epochs')
    plt.legend()
    plt.show()
    
    plt.figure(figsize=(10, 8))
    plt.plot(metrics.index,metrics.Train_jds, label='Train')
    plt.plot(metrics.index,metrics.Val_jds,label='Validation')
    plt.ylabel('JDS_metrics')
    plt.xlabel('Epochs')
    plt.legend()
    plt.show()    
    
    plt.figure(figsize=(10, 8))
    plt.plot(metrics.index,metrics.Train_bgrej, label='Train')
    plt.plot(metrics.index,metrics.Val_bgrej,label='Validation')
    plt.ylabel('1/BKrej')
    plt.xlabel('Epochs')
    plt.legend()
    plt.show()  
    
    plt.figure(figsize=(10, 8))
    plt.plot(metrics.index,metrics.Train_bgrej-1/metrics.Train_jds, label='Train')
    plt.plot(metrics.index,metrics.Val_bgrej--1/metrics.Val_jds,label='Validation')
    plt.ylabel('1/BKrej - 1/JDS')
    plt.xlabel('Epochs')
    plt.legend()
    plt.show()    
```

```python
plot_metrics("losses_class_neg_lambda10_jsd_1511-0154.txt","Loss")
```
